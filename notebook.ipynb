### Pertanyaan 1: - Bagaimana pola temporal dan spasial polutan udara (PM2.5, PM10, SO2, NO2, CO, O3) di berbagai stasiun pemantauan?

```python
# Load data
data = pd.read_csv('dashboard/main_data.csv')

# Calculate the mean concentration of pollutants by station
mean_pollutants_by_station = data.groupby('station')[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()

mean_pollutants_by_station.reset_index(inplace=True)

plt.figure(figsize=(12, 6))
mean_pollutants_by_station.set_index('station').plot(kind='bar', stacked=False, colormap='viridis')
plt.title('Rata-rata Konsentrasi Polutan Udara Berdasarkan Stasiun Pemantauan')
plt.xlabel('Stasiun Pemantauan')
plt.ylabel('Konsentrasi (µg/m³)')
plt.xticks(rotation=45)
plt.legend(title='Polutan')
plt.tight_layout()
plt.show()
```

Visualisasi tersebut menggambarkan rata-rata konsentrasi polutan (PM2.5, PM10, SO2, NO2, CO, O3) untuk setiap stasiun pemantauan. Ini memungkinkan untuk melihat variasi kualitas udara di berbagai stasiun.

Visualisasi ini menunjukkan rata-rata konsentrasi polutan untuk setiap bulan. Ini memungkinkan kita untuk menganalisis pola temporal dan melihat bulan mana yang memiliki konsentrasi polutan yang lebih tinggi.

### Pertanyaan 2: Apakah terdapat perbedaan signifikan dalam tingkat polusi udara antara weekdays (hari kerja) dan weekends (akhir pekan) atau antar musim?

```python
# Load data
data = pd.read_csv('dashboard/main_data.csv')

# Buat kolom untuk menentukan apakah hari adalah weekday atau weekend
data['date'] = pd.to_datetime(data[['year', 'month', 'day']])
data['day_of_week'] = data['date'].dt.day_name()  # Mendapatkan nama hari
data['week_category'] = data['day_of_week'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')

# Hitung rata-rata konsentrasi polutan berdasarkan kategori
mean_pollution_week = data.groupby('week_category')[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()

plt.figure(figsize=(10, 6))
mean_pollution_week.plot(kind='bar')
plt.title('Rata-rata Konsentrasi Polutan Udara: Weekdays vs Weekends')
plt.xlabel('Kategori Hari')
plt.ylabel('Konsentrasi Polutan (µg/m³ / mg/m³)')
plt.xticks(rotation=0)  
plt.legend(title='Polutan')
plt.grid(axis='y') 
plt.tight_layout() 
plt.show()
```

Dari visualisasi tersebut, terlihat bahwa rata-rata konsentrasi PM2.5, PM10, dan O3 cenderung lebih tinggi pada akhir pekan (weekends) dibandingkan dengan hari kerja (weekdays). Hal ini menunjukkan adanya peningkatan polusi pada akhir pekan, yang bisa disebabkan oleh aktivitas manusia yang lebih intensif seperti penggunaan kendaraan pribadi dan kegiatan rekreasi.

```python
# Menentukan musim berdasarkan bulan
def determine_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

data['season'] = data['month'].apply(determine_season)

# Hitung rata-rata konsentrasi polutan berdasarkan musim
mean_pollution_season = data.groupby('season')[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()

# Visualisasi
plt.figure(figsize=(10, 6))
mean_pollution_season.plot(kind='bar')  # Hapus parameter marker
plt.title('Rata-rata Konsentrasi Polutan Udara Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Konsentrasi (µg/m³ atau mg/m³)')
plt.xticks(rotation=0)
plt.legend(title='Polutan')
plt.tight_layout()
plt.show()
```

Visualisasi kedua menunjukkan bahwa tingkat polusi udara, terutama PM2.5 dan PM10, lebih tinggi pada musim semi dan gugur dingin dibandingkan dengan musim panas dan dingin. Selain itu, konsentrasi SO2 dan NO2 juga menunjukkan pola serupa, kemungkinan disebabkan oleh peningkatan penggunaan bahan bakar fosil selama musim dingin. Sebaliknya, O3 lebih tinggi pada musim panas, yang berkaitan dengan kondisi cuaca dan interaksi sinar matahari dengan polutan prekursor O3.

## Conclusion

- Conclution pertanyaan 1: Bagaimana pola temporal dan spasial polutan udara (PM2.5, PM10, SO2, NO2, CO, O3) di berbagai stasiun pemantauan?


  Analisis polutan udara (PM2.5, PM10, SO2, NO2, CO, O3) menunjukkan bahwa kualitas udara bervariasi baik berdasarkan lokasi maupun waktu. Di stasiun Aotizhongxin, rata-rata PM2.5 mencapai 72.43 μg/m³, tertinggi dibandingkan dengan stasiun lain seperti Dingling, yang hanya 60.55 μg/m³. Hal ini menunjukkan bahwa faktor lokal seperti aktivitas industri dan kepadatan penduduk memengaruhi kualitas udara.

  Dari segi waktu, terdapat bulan-bulan tertentu dengan polusi lebih tinggi. Misalnya, Maret 2014 memiliki rata-rata PM2.5 tertinggi, 106.51 μg/m³, sedangkan Januari 2014 menunjukkan angka tinggi untuk NO2 dan SO2, kemungkinan karena aktivitas industri dan kondisi cuaca.

  Musim juga berpengaruh, dengan polusi cenderung lebih tinggi di bulan dingin, seperti Januari dan Februari, akibat penggunaan bahan bakar untuk pemanas dan inversi suhu yang menyebabkan penumpukan polusi.

  Analisis lokasi menunjukkan bahwa stasiun di daerah perkotaan, seperti Dongsi dan Guanyuan, mengalami polusi lebih tinggi dibandingkan stasiun di daerah pinggiran seperti Changping. Ini menunjukkan bahwa aktivitas manusia memiliki dampak signifikan terhadap kualitas udara.
  
- Conclution pertanyaan 2: Apakah terdapat perbedaan signifikan dalam tingkat polusi udara antara weekdays (hari kerja) dan weekends (akhir pekan) atau antar musim?


  Terdapat perbedaan yang signifikan dalam tingkat polusi udara antara hari kerja (weekdays) dan akhir pekan (weekends). Pada akhir pekan, konsentrasi polutan seperti PM2.5 (73.48 µg/m³), PM10 (96.05 µg/m³), SO2 (15.23 µg/m³), dan O3 (59.25 µg/m³) cenderung lebih tinggi dibandingkan dengan hari kerja, di mana konsentrasi PM2.5 tercatat 68.90 µg/m³ dan PM10 91.39 µg/m³. Peningkatan ini mungkin disebabkan oleh perubahan aktivitas manusia pada akhir pekan, seperti peningkatan penggunaan transportasi pribadi dan kegiatan rekreasi.

  Dari segi musiman, polusi udara juga bervariasi. PM2.5 dan PM10 mencapai puncaknya pada bulan Maret, sementara konsentrasi SO2 dan NO2 lebih tinggi pada bulan Januari. Hal ini menunjukkan bahwa musim dingin dan semi cenderung memiliki polusi yang lebih tinggi, kemungkinan akibat dari penggunaan pemanas dan aktivitas industri yang lebih intensif. Di sisi lain, konsentrasi ozon (O3) lebih tinggi selama musim panas, terkait dengan interaksi antara sinar matahari dan polutan lainnya.

- Conclution pertanyaan 3: Bagaimana pengaruh arah dan kecepatan angin terhadap penyebaran polutan udara?

  Pengaruh arah dan kecepatan angin terhadap penyebaran polutan udara sangat signifikan. Kecepatan angin yang rendah, terutama di arah timur, berhubungan dengan konsentrasi polutan yang lebih tinggi seperti PM2.5, PM10, NO2, dan SO2. Hal ini menunjukkan bahwa pada kecepatan angin rendah, polusi cenderung terperangkap dan tidak dapat menyebar, menyebabkan peningkatan konsentrasi polutan di wilayah tersebut.

  Sebaliknya, pada kecepatan angin yang lebih tinggi, terutama dari arah barat daya, konsentrasi partikel polutan seperti PM2.5 dan PM10 cenderung menurun. Namun, konsentrasi ozon (O3) mengalami peningkatan yang signifikan pada kecepatan angin tinggi. Ini menunjukkan bahwa angin dengan kecepatan tinggi dapat mempercepat pembentukan ozon melalui reaksi kimia yang didorong oleh interaksi dengan sinar matahari.

  Secara keseluruhan, kecepatan angin rendah berhubungan dengan tingginya konsentrasi polutan partikulat, sedangkan kecepatan angin tinggi dapat menurunkan konsentrasi polutan tetapi meningkatkan pembentukan ozon. Arah angin juga berperan penting, di mana arah timur menunjukkan polusi yang lebih tinggi, sementara arah barat daya berkorelasi dengan peningkatan konsentrasi ozon pada kecepatan angin yang lebih tinggi.
